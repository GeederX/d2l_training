### Imports ###

```python
import torch
import torchvision
import os
from torch.utils import data
from torchvision import transforms
```

### Hyper parameters ###

```python
batch_size = 256
lr = 0.1
num_epochs = 5

num_inputs = 784
num_outputs = 10
```

### Parameters Init ###

```python
W = torch.normal(0, 0.01, size=(num_inputs, num_outputs), requires_grad=True)
b = torch.zeros((num_outputs), requires_grad=True)
```

### Load  ###

```python
def load_data_fashin_mnist(batch_size, resize=None, num_threads=0):
    trans = [transforms.ToTensor()]
    if resize:
        trans.insert(0, transforms.Resize(resize))
    trans = transforms.Compose(trans)
    mnist_train = torchvision.datasets.FashionMNIST(
        root=os.path.join(".", "data"), train=True, transform=trans, download=True
    )
    mnist_test = torchvision.datasets.FashionMNIST(
        root=os.path.join(".", "data"), train=False, transform=trans, download=True
    )
    return (
        data.DataLoader(mnist_train, batch_size, shuffle=True, num_workers=num_threads),
        data.DataLoader(mnist_test, batch_size, shuffle=False, num_workers=num_threads),
    )
```

### Softmax ###

```python
def softmax(X):
    X_exp = torch.exp(X)
    partition = X_exp.sum(dim=-1, keepdim=True)
    return X_exp / partition
```

### Cross entropy ###

```python
def cross_entropy(y, y_hat):
    return -torch.log(y_hat[range(len(y_hat)), y])
```

### Accuracy Asess ###

```python
def accuracy(y, y_hat):
    y_hat = y_hat.argmax(axis=-1)
    cmp = y_hat.type(y.dtype) == y
    return float((cmp.type(y.dtype).sum()) / len(y))
```

### Net ###

```python
def net(X):
    return softmax(torch.matmul(X.reshape(-1, W.shape[0]), W) + b)
```

### Debug ###

```python
TrainData, TestData = load_data_fashin_mnist(batch_size, resize=None, num_threads=0)
temp = next(iter(TrainData))
X = temp[0].reshape(batch_size, -1)
y = temp[1]
y_hat = torch.matmul(X, W) + b
y_hat = softmax(y_hat)
accuracy(y, y_hat)
```
